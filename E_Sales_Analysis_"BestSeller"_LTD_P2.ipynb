```python
!pip install google-cloud-bigquery

from scipy.stats import shapiro
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import scipy.stats as stats

%matplotlib inline
pd.set_option('display.width', 1000)
pd.options.display.float_format = "{:.5f}".format

# Аутентифікація в Google Big Querry
from google.colab import auth
auth.authenticate_user()

from google.cloud import bigquery

# 1) Ініціалізуємо клієнт
client = bigquery.Client(
    project="data-analytics-mate"
)

# 2) SQL-запит.
query = """
SELECT
      s.date,
      s.ga_session_id,
      sp.continent,
      sp.country,
      sp.device,
      sp.browser,
      sp.mobile_model_name,
      sp.operating_system,
      sp.language,
      sp.medium,
      sp.channel,
      a.id AS account_id,
      a.is_verified,
      a.is_unsubscribed,
      p.category,
      p.name,
      p.price,
      p.short_description
FROM `DA.session` s
JOIN `DA.session_params` sp ON s.ga_session_id = sp.ga_session_id
LEFT JOIN `DA.account_session` acs ON s.ga_session_id = acs.ga_session_id
LEFT JOIN `DA.account` a ON acs.account_id = a.id
LEFT JOIN `DA.order` o ON s.ga_session_id = o.ga_session_id
LEFT JOIN `DA.product` p ON o.item_id = p.item_id
"""

# 3) Запускаємо запит
query_job = client.query(query)

# 4) Отримуємо результат і перетворюємо в DataFrame
results = query_job.result()
df = results.to_dataframe()
```

### **6. Різниця поведінки (у плані продажів) тих, хто відписався від розсилки та тих, хто  підписаний**


```python
# Фільтруємо реальні покупки (видаляємо NaN та/або ціни <= 0)
df_sales = df.dropna(subset=['price'])
df_sales = df_sales[df_sales['price'] > 0]

# Групуємо за account_id, щоб отримати суму покупок по кожному користувачу
# або 'min', якщо значення завжди однакове для одного account_id
df_users = (df_sales.groupby('account_id', as_index=False).agg({'price': 'sum', 'is_unsubscribed': 'max'}))

# Тепер у df_users маємо:
# account_id | price (сума покупок) | is_unsubscribed (True/False)

#Поділ на 2 групи
df_unsub = df_users[df_users['is_unsubscribed'] == 1]['price']
df_sub = df_users[df_users['is_unsubscribed'] == 0]['price']

print("Кількість відписаних користувачів:", len(df_unsub))
print("Кількість підписаних користувачів:", len(df_sub))

# Побудуємо гістограми/щільності
plt.figure(figsize=(10, 4))

sns.kdeplot(df_unsub, fill=True, label='Відписані')
sns.kdeplot(df_sub, fill=True, label='Підписані')

plt.title('Розподіл сум покупок (price) у двох групах')
plt.xlabel('Сума покупок')
plt.ylabel('Щільність')
plt.legend()
plt.show()
```

###**9. Продажі за топ-10 категоріями та топ-5 країнами**

```python
# 1. Групуємо за категоріями та сумуємо продажі
df_top_categories = (
    df.groupby('category', as_index=False)['price']
      .sum()
      .sort_values('price', ascending=False)
      .head(10)
)

# 2. Групуємо за країнами та сумуємо продажі
df_top_countries = (
    df.groupby('country', as_index=False)['price']
      .sum()
      .sort_values('price', ascending=False)
      .head(5)
)

# Перетворимо ці дані у списки, щоб зручно використати для фільтрації
top_10_categories_list = df_top_categories['category'].tolist()
top_5_countries_list = df_top_countries['country'].tolist()

# Фільтруємо вихідний DataFrame, щоб залишити лише топ-10 категорій та топ-5 країн
df_filtered = df[
    df['category'].isin(top_10_categories_list) &
    df['country'].isin(top_5_countries_list)
]

df_filtered = df_filtered.sort_values(by="country", ascending=True)

#Pviot table
df_pivot_sales = pd.pivot_table(
    data=df_filtered,
    index='category',   # Рядки - категорії
    columns='country',  # Стовпці - країни
    values='price',     # Що агрегіюємо
    aggfunc='sum'       # Сума продажів
)
# Пояснення:
# index='category': кожен рядок таблиці буде відповідати конкретній категорії з топ-10.
# columns='country': кожен стовпець — країна з топ-5.
# values='price': ми хочемо підсумовувати продажі (колонка price).
# aggfunc='sum': застосовуємо сумування.

#Сортуємо від більшого до меншего
df_pivot_sales = df_pivot_sales.sort_values(by="united states", ascending=False)
# 1. Обчислюємо суму продажів по кожній країні (стовпчику)
totals = df_pivot_sales.sum(axis=0)
# 2. Сортуємо імена стовпців за сумою продажів у порядку убування
sorted_cols = totals.sort_values(ascending=False).index.tolist()
# 3. Реорганізуйте зведену таблицю відповідно до нового порядку стовпців
df_pivot_sorted = df_pivot_sales[sorted_cols]
df_pivot_sorted
```

##**Статистичний аналіз**

### **10. Кореляція між кількістю сесій та продажами:**
Коефіцієнт кореляції Пірсона ≈ 0.79, що свідчить про досить сильний позитивний зв’язок.

p-значення ≈ 6.48e-21 (надзвичайно мале), тобто кореляція статистично значуща з великою ймовірністю (навіть при α=0.001).

Кореляція «кількість сесій — сума продажів». Дані свідчать про сильний позитивний зв’язок (r ~ 0.79). Це хороший знак: збільшуючи трафік (наприклад, через маркетинг чи SEO), ми можемо розраховувати на зростання продажів.

```python
##Statistics
df_daily = df.groupby('date').agg({
    'ga_session_id': 'nunique',  # кількість унікальних сесій
    'price': 'sum'               # сума продажів
}).reset_index()

# Обчислюємо коефіцієнт кореляції (Пірсона) та p-значення:
corr_coef, p_value = stats.pearsonr(df_daily['ga_session_id'], df_daily['price'])

print("Коефіцієнт кореляції (Pearson):", corr_coef)
print("p-значення:", p_value)
# Analyze statistical significance
if p_value < 0.05:
    print("Кореляція є статистично значущою.")
else:
    print("Кореляція НЕ є статистично значущою.")

# scatter plot з регресійною лінією
sns.set(style="whitegrid")
plt.figure(figsize=(9, 5))
sns.regplot(x='ga_session_id', y='price', data=df_daily, scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title("Correlation between the number of sessions and sales")
plt.xlabel("Number of unique sessions")
plt.ylabel("Sales amount $")
plt.show()
```

###**11. Відмінності у продажах зареєстрованих та незареєстрованих користувачів (t-тест)**
t-статистика ≈ -1.03,
p-значення ≈ 0.30.

Boxplot свідчить, що розподіли дуже схожі й містять багато викидів (крапок вище «вусів»).
Значна частина користувачів (у обох групах) має низькі або близькі витрати, тоді як невелика кількість покупців у кожній групі може витрачати значно більше (дивлячись на точки вище 2000, 4000 тощо).

- Порівняння «зареєстровані vs. незареєстровані». За результатами t-тесту (p≈0.30) немає підстав вважати, що зареєстровані користувачі витрачають більше чи менше, ніж незареєстровані, з погляду середньої суми покупок.

- Boxplot показує великий розкид і наявність викидів в обох групах, що типово для даних про продажі, де частина клієнтів роблять дуже дорогі покупки.

```python
# Створюємо дві групи: зареєстровані та незареєстровані користувачі.
# Припустимо, що якщо account_id є, користувач зареєстрований.
registered = df[df['account_id'].notna()]['price']
unregistered = df[df['account_id'].isna()]['price']

# Видаляємо NaN-значення з колонок 'price'
registered = registered.dropna()
unregistered = unregistered.dropna()

# Проведемо незалежний t-тест (Welch's t-test)
t_stat, p_val = stats.ttest_ind(registered, unregistered, equal_var=False)

print("t-статистика:", t_stat)
print("p-значення:", p_val)

# Створимо новий стовпець, який показує, чи користувач зареєстрований (True) чи ні (False)
df['is_registered'] = df['account_id'].notna()

# Відфільтруємо дані по продажах (видаляємо NaN)
df_sales = df[['is_registered', 'price']].dropna()

# Побудова boxplot за допомогою seaborn
plt.figure(figsize=(8,6))
sns.boxplot(x='is_registered', y='price', data=df_sales)
plt.xticks([0, 1], ['Незареєстровані', 'Зареєстровані'])
plt.title("Порівняння продажів: зареєстровані vs незареєстровані користувачі")
plt.xlabel("Статус користувача")
plt.ylabel("Сума продажів")
plt.show()
```
